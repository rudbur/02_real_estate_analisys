## Предобработка данных

### Пропуски

В столбцах `locality_name` и `floors_total` меньше 0.5% пропусков. К тому же анализировать объявления с неопределенным населённым пунктом - лишнее. Поэтому просто удалим их.  
Пропуск в `floors_total` может означать одноэтажность постройки - пользователь просто не стал утруждать себя вводом данных.  Пропуск в `locality_name`, возможно, из-за приложения, через которое размещалось объявление, - оно не смогло определить геолокацию.

```python
df.dropna(subset=['floors_total', 'locality_name'], inplace=True)
```

В `living_area` и `kitchen_area` пропуски заполним через отношение общей площади к ним. В `ceiling_height` - медианным значением.  
Причиной пропуска вполне может быть незнание этого параметра пользователем - люди зачастую ориентируются на общую площадь.

```python
fillna_living = (
    df.groupby('rooms').living_area.transform('median') / 
    df.groupby('rooms').total_area.transform('median') * df.total_area
)
```

```python
df.living_area.fillna(fillna_living, inplace=True)
```

```python
fillna_kitchen = (
    df.groupby('rooms').kitchen_area.transform('median') / 
    df.groupby('rooms').total_area.transform('median') * df.total_area
)
```

```python
df.kitchen_area.fillna(fillna_kitchen, inplace=True)
```

```python
df.ceiling_height.fillna(df.ceiling_height.median(), inplace=True)
```

А вот с количеством дней до продажи не так ясно: это может быть и сбой системы и указывать на то, что квартира всё ещё не продана.

```python
df[df.first_day_exposition > '2019-00-01T00:00:00'].days_exposition.isna().sum() /\
df.days_exposition.isna().sum()
```

Число пропущенных значений в `days_exposition` в последние полгода составляет больше половины от всех пропущенных значений в этом столбце. Гипотеза о том, что квартиру не успели продать, звучит правдоподобно. Пропуски заполнять не будем.

В `is_apartment` пропуски означают, что продаваемая недвижимость апартаментами скорее всего не является, потому заполнять их будем значением `False`.  
В `balcony` пропуски скорее всего означают отсутствие балкона, потому заполним их значением `0`.

```python
df.is_apartment.fillna(False, inplace=True)
df.balcony.fillna(0, inplace=True)
```

Остальные пропуски оставим в покое.

```python
df.isna().mean()
```

### Дубликаты

Пройдёмся по значениям `locality_name`, посмотрим, есть ли там скрытые дубликаты - похожие названия населённых пунктов.

```python
list(df.locality_name.sort_values(ascending=True).unique())
```

Видим такие дубликаты, как:
- посЕлок и посЁлок
- разные типы посёлков(городской, городского типа, коттеджный, при железнодорожной станци)
- садоводческое некоммерческое товарищество и садовое товарищество

Приведём это всё к единому виду.

```python
#любые посёлки сделаем просто посёлками
localities_to_replace = ['поселок','городской посёлок','посёлок городского типа', 'коттеджный посёлок', \
                        'посёлок при железнодорожной станции', 'посёлок станции']
for locality in localities_to_replace:
    df.locality_name = df.locality_name.str.replace(locality, 'посёлок')
    
#садовое некоммерческое превратим в садовое    
df.locality_name = df.locality_name.str.replace('садоводческое некоммерческое товарищество', 'садовое товарищество')
```

```python
localities = list(df.locality_name.sort_values().unique())
localities
```

Мы получили следующие типы населённых пунктов:
- город (слово не значится в названии)
- деревня
- посёлок
- садовое товарищество
- село  

Может такое быть, что населённый пункт с необозначенным типом дублируется. Проверим это.

```python
#проверим, встречаются ли населённые пункты без типа в других строках
localities_similar = {}
first_town_index = localities.index('деревня Агалатово')
for city in localities[:first_town_index]:
    for town in localities[first_town_index:]:
        if city in town:
            localities_similar[city] = town
localities_similar
```

Найдя каждый населённый пункт на карте, выясним, что Любань, Мурино, Кудрово - интересующие нас скрытые дубликаты. Приведём их к единому виду.

```python
df.locality_name = df.locality_name.str.replace('деревня Кудрово', 'Кудрово')\
.replace('посёлок Любань' 'Любань').replace('посёлок Мурино', 'Мурино')
```

### Аномалии

С аномальными значениями в основном будет работать позже, когда приступим к исследовательскому анализу. Здесь же обратим внимание на столбец `balcony` и `ceiling_height`.

```python
df.balcony.unique()
```

Больше двух балконов в квартире - маловероятно. В этих случаях сделаем замену указанного количества на медианное значение.

```python
df.loc[df.balcony > 2, 'balcony'] = df[df.balcony>0].balcony.median()
```

Отобразим высоту потолков.

```python
df.ceiling_height.sort_values(ascending=False).unique()
```

- Высота потолков меньше 2м - маловероятна, как и высота от 5 до 20м. В этих случаях заменим значения медианой.  
- В значениях от 22 до 32 скорее всего была допущена ошибка ввода точки, разделяющей целую и дробную часть. В этих случаях значения поделим на 10.  
- 100м - заменим медианой. 

```python
df['ceiling_height'][((df.ceiling_height<=20) & (df.ceiling_height>5)) | (df.ceiling_height<=2) |\
                     (df.ceiling_height>50)] = df.ceiling_height.median()
df['ceiling_height'][(df.ceiling_height>=22) & (df.ceiling_height<=50)] = df.ceiling_height/10
```

Округлим высоту до одного знака после запятой. Точность до 1см здесь лишняя, и заодно проведём группировку данных по дециметрам.

```python
df.ceiling_height = (df.ceiling_height*10).astype(int)/10
df.ceiling_height.sort_values(ascending=False).unique()
```

### Фильтрация датафрейма

Уберём из нашего датафрейма редкие, выбивающиеся значения, основываясь на теории диаграмм размаха.

```python
#чтобы сохранить больше данных, границами выберем 20% и 80%
#функция возвращает список булевых значений, основывающийся на поданном в неё столбце датафрейма
def filter_data(column):
    Q1 = column.quantile(0.20)
    Q3 = column.quantile(0.80)
    IQR = Q3 - Q1
    filter = (column >= Q1 - 1.5 * IQR) & (column <= Q3 + 1.5 *IQR)
    return filter
```

```python
#пройдёмся по интересующим нас столбцам и получим итоговый булев список, описывающий,
#какие строки датафрейма сохранить, какие - удалить.
filter = filter_data(df.last_price)
for column in columns_filtered:
    filter = filter & filter_data(df[column])

df_good = df[filter]
```

```python
#группируем, считаем среднее и строим график
def plot(data, x, y, kind='line'):
    data.pivot_table(index = x, values=y, aggfunc='mean').plot(y=y, kind=kind, grid=True, style='o-')
    return plt.show()
```

```python
#группируем, считаем среднее и вычисляем отношение между минимальным и максимальным значением в процентах
def minmax_ratio(df, column):
    ratio = 1 - df.pivot_table(index=column, values='square_price', aggfunc='mean')['square_price'].min() /\
    df.pivot_table(index=column, values='square_price', aggfunc='mean')['square_price'].max()
    ratio = int(ratio * 1000)/10
    return ratio
```

```python
minmax_ratio(df_good, 'floor_is')
```
